### Lead Time Calculation

```python
import math
import numpy as np

# point a
a1 = math.radians(0) # lat
a2 = math.radians(140) # long
a3 = 0 # in meters

# point b
b1 =  math.radians(0) # lat
b2 =  math.radians(-140) # long
b3 = 0 # in meters

r = 6371000 #in meters the Earth's Radius

# calculate Cartesian Coordinates
x_a = (r + a3)*math.cos(a1)*math.cos(a2)
y_a = (r + a3)*math.cos(a1)*math.sin(a2)
z_a = (r + a3)*math.sin(a1)

x_b = (r + b3)*math.cos(b1)*math.cos(b2)
y_b = (r + b3)*math.cos(b1)*math.sin(b2)
z_b = (r + b3)*math.sin(b1)

# distance between two points
d = math.sqrt((x_b - x_a)**2 + (y_b - y_a)**2 + (z_b - z_a)**2) #in meters

# AR airspeed
v =   275  #mph
v_m =  v * 1609.34 #meters per hour

# Time it takes to travel

t = d/v_m 

print(f'It takes {t} hours for tropical Pacific air to travel close to North American Coast')
```

### Import CW3E collected radiosonde data

```python
#### Vapor Transport Calculation Test
t = 20 # actual temp 8
td = 12.5 # dew point temp 9
w =  4# wind speed in m/s 11

vp = 6.112 * math.exp((17.67 * td)/(td + 243.5)) # td is dew point temperature in degrees celsius and vp in hPa
a_humidity = 216.7 * vp/(t + 273.15) # t is actual temperature in celsius and a_humidity will be in grams/m**3
mt = w * a_humidity/1000 # vapor transport is in kg/m**2/s

print(f'Vapor transport is {mt} kg/m**2/s')
```

```python
#### define vapor transport function
def vaportranspor(t, td, w):
    vp = 6.112 * math.exp((17.67 * td)/(td + 243.5)) # td is dew point temperature in degrees celsius and vp in hPa
    a_humidity = 216.7 * vp/(t + 273.15) # t is actual temperature in celsius and a_humidity will be in grams/m**3
    mt = w * a_humidity/1000 # vapor transport is in kg/m**2/s
    return(mt)
```
