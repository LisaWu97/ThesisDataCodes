```python
import pandas as pd

#### SOI
soi = pd.read_csv(r"C:\Users\lisaw\OneDrive\Documents\temp\ThesisRepo\Data\Indexes\soi.long.csv") # monthly
soi.set_index('Date', inplace = True)
# print(soi)

#### PDO
pdo = pd.read_csv(r"C:\Users\lisaw\OneDrive\Documents\temp\ThesisRepo\Data\Indexes\pdo.timeseries.sstens.csv") # monthly
pdo.set_index('Date', inplace = True)
# print(pdo)

#### AO
ao = pd.read_csv(r"C:\Users\lisaw\OneDrive\Documents\temp\ThesisRepo\Data\Indexes\ao.long.csv") # monthly
ao.set_index('Date', inplace = True)
# print(ao)

#### MJO 
mjo = pd.read_csv(r"https://psl.noaa.gov/mjo/mjoindex/romi.cpcolr.1x.txt", sep = "\s+", header = None) # daily data
mjo.columns = ['Year','Month','Day','Hour','PC1','PC2','PC1_2_Amp']
mjo['Date'] = pd.to_datetime(mjo[['Year', 'Month', 'Day']])
mjo.set_index('Date', inplace = True)
print(mjo)
```

```python
#### join soi, pdo and ao (same temporal resolution)
import numpy as np

s_p_a = soi.join([pdo, ao], how = 'outer')
s_p_a.columns = ['SOI', 'PDO', 'AO']
s_p_a.index = pd.to_datetime(s_p_a.index)
s_p_a_m = s_p_a.join(mjo, how = 'outer')
s_p_a_m['Year'] = s_p_a_m.index.year
s_p_a_m['Month'] = s_p_a_m.index.month
s_p_a_m['Day'] = s_p_a_m.index.day
s_p_a_m['Hour'] = np.nan
print(s_p_a_m)
```
